```python
import pandas as pd

"""
使用pandas的read_csv函数，
读取breast_cancer.csv文件，
并将文件数据赋值给变量df。
"""
df = pd.read_csv('../datasets/breast_cancer/breast_cancer.csv')

# 查看df
df
```

```python
# 用具体的列名选择df的三列，构成df_part。
df_part = df[['diagnosis', 'radius_mean', 'perimeter_mean']]

# 查看df_part
df_part
```

```python
"""
将df_part的数据，
按照csv的格式写入到breast_cancer_part.csv文件。
"""
df_part.to_csv('../datasets/breast_cancer/breast_cancer_part.csv', index=False)

"""
再次使用pandas的read_csv函数，
读取breast_cancer_part.csv文件，
并将文件数据赋值给变量df_new。
"""
df_new = pd.read_csv('../datasets/breast_cancer/breast_cancer_part.csv')

# 校验df_new与df_part是否一致。
df_new
```
