```python
import pandas as pd

"""
使用pandas的read_json函数，
读取iris.json文件，
并将文件数据赋值给变量df。
"""
df = pd.read_json('../datasets/iris/iris.json')

# 查看df
df
```

```python
# 选择df的前10行数据，构成df_part
df_part = df[:10]

# 查看df_part
df_part
```

```python
"""
将df_part的数据，
按照json的格式写入到iris_part.json文件。
"""
df_part.to_json('../datasets/iris/iris_part.json')

"""
再次使用pandas的read_json函数，
读取iris_part.json文件，
并将文件数据赋值给变量df_new。
"""
df_new = pd.read_json('../datasets/iris/iris_part.json')

# 校验df_new与df_part是否一致。
df_new
```
