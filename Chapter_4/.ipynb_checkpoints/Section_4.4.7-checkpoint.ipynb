```python
import pandas as pd


df = pd.read_excel('../datasets/bitcoin/bitcoin.xlsx', index_col='Date')

df
```

```python
def cal_change_rate(x):
    '''
    定义一个函数，计算当日股票价格的增长率。
    '''
    return (x['Close'] - x['Open']) / x['Open']
```

```python
#使用自定义函数，操作数据框。
df.apply(cal_change_rate, axis=1).sort_values()
```
